```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "NIMO_target_list.csv",
    dtype={"gaiadr3_source_id": "string", "TIC": "string"}
)
df.head()
```

```python
df['# of sectors'].sum()
```

```python
mask = df["gaiadr3_source_id"].notna() & df["gaiadr3_source_id"].str.strip().ne("")
df.loc[mask, "# of sectors"].sum()
```

```python
df = df.drop(columns=["other_id", "srmper"])
df.head()
```

```python
id_cols = df.columns[:5].tolist()
sector_cols = [c for c in df.columns if c.startswith("Sector")]

df_rotated = (
    df.melt(
        id_vars=id_cols,
        value_vars=sector_cols,
        var_name="Sector",
        value_name="name"
    )
    .dropna(subset=["name"])
    .loc[lambda x: x["name"].astype(str).str.strip() != ""]
    .copy()
 )

df_rotated["Sector"] = (
    df_rotated["Sector"]
    .str.replace("Sector", "", regex=False)
    .str.strip()
 )
```

```python
df_rotated[df_rotated.nimo_id==2]
```

```python
df_rotated["lc_name"] = df_rotated.apply(
    lambda row: f"TIC-{row['TIC']}-{row['Sector']}-{row['name']}", 
    axis=1
)
```

```python
df_rotated.head(20)
```

```python
df_rotated = df_rotated.rename(columns={'# of sectors': 'num_sectors'})
```

```python
df_rotated.to_csv('NIMO_lcnames.csv', index=False)
```

```python
df_rotated[df_rotated['gaiadr3_source_id'].notna()].to_csv('NIMO_gaiaids.csv', index=False)
```
