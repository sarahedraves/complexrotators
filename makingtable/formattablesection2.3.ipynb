```python
table=table[table['pop_id'].notnull()]
```

```python
columns=['pop_id',
         'g_absmag','phot_g_mean_mag','phot_g_mean_flux_over_error',
         'bp_absmag','phot_bp_mean_mag','phot_bp_mean_flux_over_error',
         'rp_absmag','phot_rp_mean_mag','phot_rp_mean_flux_over_error',
         'ra','ra_error','dec','dec_error','parallax','parallax_error','ruwe']
```

```python
table=table[columns]
```

```python
rawtable=pd.merge(ids,table,how='left',on='pop_id')
```

```python
rawtable.drop(['pop_id','disco_paper'],axis=1,inplace=True)
```

```python
rawtable.rename(columns={'author':'discovery_paper_author',
                         'year':'discovery_paper_year'},inplace=True)
```

```python
rawtable['per']=rawtable.per.round(4)
rawtable['per2']=rawtable.per2.round(4)
```

```python
rawtable.to_csv('papertable2.3.csv',index=False)
```

```python
rawtable
```

```python
#first set number of sig figs for each 
printtable=rawtable.drop(['phot_g_mean_flux_over_error','phot_bp_mean_flux_over_error',
                         'phot_rp_mean_flux_over_error','ra_error','dec_error',
                         'parallax_error','g_absmag','bp_absmag','rp_absmag',
                         'gaiadr3_source_id','epic_id'],axis=1)
```

```python
printtable=printtable.groupby(['discovery_paper_author','discovery_paper_year']).sample(n=1)
```

```python
printtable.sort_values(by=['discovery_paper_year','discovery_paper_author'],inplace=True)
```
