```python
# Convert columns to sets
set1 = set(filteredids["pop_id"].dropna())
set2 = set(filteredlcs["pop_id"].dropna())

# Find values unique to each
only_in_col1 = set1 - set2
only_in_col2 = set2 - set1

print(f"Values only in col1: {only_in_col1}")
print(f"Values only in col2: {only_in_col2}") #figuring out which object has an EPIC ID but no campaign
```

```python
# Count non-NaNs per row
notna_counts = bestlcs[campaign_cols].notna().sum(axis=1)

# Create mask for rows with 2 or more non-NaNs
mask3 = notna_counts >= 2

# Filter DataFrame
filtered2 = bestlcs[mask3]
filtered2 #figuring out which objects appear in more than one campaign
```
